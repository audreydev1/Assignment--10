# 1.0 Importing Python Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

# 1.1 Loading the dataset


```
conversion = pd.read_csv('KAG_conversion_data.csv')
conversion
```

```
conversion.info()
```

```
conversion.describe()
```

# 1.2 Dummy encoding categorical or object values

```
X = pd.get_dummies(data = conversion, columns = ['age', 'gender'])
X
```

# 1.4 Using a heat map to show the correlation in the data.

```
X.corr()
```

```
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
sns.set_style("dark")
sns.heatmap(conversion.corr())
```

# 1.5 Dropping the first 4 columns in the data frame X.

```
X = X.drop(['ad_id', 'xyz_campaign_id', 'fb_campaign_id','interest'], axis = 1)
X
```

### We dropped the first three columns('ad_id', 'xyz_campaign_id', 'fb_campaign_id') because they are IDs and we cannot learn anything from them because they don't vary in addition to being ,the fourth column is dropped because it has a high correlation to the ID columns which are not useful to us yet it correlates lowly with the othe variables

# 1.6 Using the elbow method to determine the best number of clusters for the data in the range of 2 to 20. 

```
conv = []
for i in range(2, 21):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(X)
    conv.append(km.inertia_)
    

plt.figure(figsize=(16,10))
plt.plot(range(2, 21), conv)
plt.title('The Elbow Curve', fontsize = 20)
plt.grid(True)
plt.xlabel('No. of Clusters')
plt.ylabel('conv')
plt.show()
```

### The Elbow method shows that the elbow point is 5 clusters

# 1.7 Using the elbow point to cluster the values in the data frame X.

```
kmeans=KMeans(n_clusters=5,random_state=0)
#fitting the model
kmeans.fit(X)
```

# 1.8 Using the model to predict the labels from the data and save them to variable y_means

```
# using the model to predict labels
y_means=kmeans.predict(X)
```

# 1.9 Adding y_means to conversion as column â€˜Advert_Typeâ€™.

```
conversion['Advert_Type'] = y_means
conversion
```

# 1.10 Which advert type lead to the highest and consistent amount of sales by customers of all the age brackets?

```
#visualizing advert type,approved conversions and age
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
sns.set_style("dark")
ax=sns.barplot(data=conversion,x='age',y='Approved_Conversion',hue='Advert_Type')
plt.show()
```

# 1.11 Does the company xyz have gender bias in terms of their ad spending? Are their products gender neutral?

```
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
sns.set_style("dark")
ax=sns.barplot(data=conversion,x='gender',y='Spent',hue='Advert_Type')
plt.show()
```

### Company xyz are gender neutral since there is no bias in terms of their ad spending

# 1.12 Perform the same analysis using hierarchical clustering and compare the results in terms of the number of useable clusters.

```
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(15, 10))
plt.title("Conversion Dendograms")
dend = shc.dendrogram(shc.linkage(X, method='ward'))
```

```
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(affinity='euclidean', linkage='ward')
cluster.fit_predict(X) 
y_hier = cluster.labels_
```

```
#Adding the hierarchical cluster
conversion['Advert_Type2'] = y_hier
conversion
```

```
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
sns.set_style("dark")
ax=sns.barplot(data=conversion,x='Advert_Type2',y='Approved_Conversion',hue='Advert_Type')
plt.show()
```

### The K-Means algorithm produced 5 useable clusters while the Hierarchical algorithm produced only 2 clusters
